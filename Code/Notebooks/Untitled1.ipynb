```python
a = 0.2
b = 0.2
c = 5.7

from jitcode import y
roessler_f = [
    -y(1) - y(2),
     y(0) + a*(1),
    b + y(2)*(y(0)-c)
]

from jitcode import jitcode

I = jitcode(roessler_f)

I.set_integrator("dopri5")

initial = [0.1, 0.2, 0.3]

I.set_initial_value(initial, time=0.0)

times = range(100, 201)
for time in times:
    print(I.integrate(time))
```
